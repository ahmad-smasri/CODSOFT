## Libraries to be Used

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Reading Training and Testing CSV File

```python
#Training Set
training_data_file = pd.read_csv("/content/drive/MyDrive/Project 2 Dataset/fraudTrain.csv")
training_data = pd.DataFrame(training_data_file)
training_data

#Testing Set
testing_data_file = pd.read_csv("/content/drive/MyDrive/Project 2 Dataset/fraudTest.csv")
testing_data = pd.DataFrame(testing_data_file)
testing_data
```

## Removing unneeded columns

```python
#Training Set
training_needed_data = training_data.drop(["Unnamed: 0", 'trans_date_trans_time', "merchant", "first", "last", "gender", "street", "city", "state",
                                           "zip", "city_pop", "job", "dob", "trans_num"], axis =1)
print(training_needed_data)

#Testing Set
testing_needed_data = testing_data.drop(["Unnamed: 0",'trans_date_trans_time', "merchant", "first", "last", "gender", "street", "city", "state",
                                         "zip", "city_pop", "job", "dob", "trans_num"], axis =1)
print(testing_needed_data)
```

## Encoding Categories


```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
le = LabelEncoder()

#Training Set
# Fit the encoder to the categorical data
le.fit(training_needed_data["category"])

# Transform the categorical data into numerical labels
encoded_cat = le.transform(training_needed_data["category"])
training_needed_data["category"] = encoded_cat

print(training_needed_data)

#Testing Set
# Fit the encoder to the categorical data
le.fit(testing_needed_data["category"])

# Transform the categorical data into numerical labels
encoded_cat = le.transform(testing_needed_data["category"])
testing_needed_data["category"] = encoded_cat

print(testing_needed_data)
```

## Preprocessing Training Data

### Distributing Data columns between features and target

```python
#Training Set
#Features
training_feature_data = training_needed_data[["unix_time", "cc_num", "category", "amt", "lat", "long", "merch_lat", "merch_long"]]
#Target
training_Fraudulence = training_needed_data[["is_fraud"]]

#Testing Set
#Features
testing_feature_data = testing_needed_data[["unix_time", "cc_num", "category", "amt", "lat", "long", "merch_lat", "merch_long"]]
#Target
testing_Fraudulence = testing_needed_data[["is_fraud"]]
```

### Preprocessing Data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#Training Set
#Transforming scaled data
training_feature_data [["cc_num", "amt"]] = scaler.fit_transform(training_feature_data [[ "cc_num", "amt"]])
training_feature_data

#Testing Set
#Transforming scaled data
testing_feature_data [["cc_num", "amt"]] = scaler.fit_transform(testing_feature_data [[ "cc_num", "amt"]])
testing_feature_data
```

## Train and Test

### Accuracy key indicator

```python
from sklearn.metrics import r2_score


def predict(dataset, target, model):
  # Predict the target column using the model
  predictions = model.predict(dataset).round(0)
  # Create a new Dataframe to compare the target values and the predicted values
  output = dataset.copy()
  # Adding the target values in a 'target' column
  output['Actual'] = target
  # Adding the predicted values in a 'prediction' column
  output['Prediction'] = predictions
  # Calculate the absolute errors
  errors = abs(output['Actual'] - output['Prediction'])
  # Calculate the mean absolute error (mae)
  loss = np.mean(errors)
  # Print out the mean absolute error (mae)
  print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
  # Calculate mean absolute percentage error (MAPE)
  mape = 100 * (errors / target.shape[0])
  # Calculate and display accuracy
  Error = 100 * loss/target.shape[0]
  print('Accuracy:', 100 - round(Error, 2), '%.')
  #R2
  R2=np.corrcoef(output['Actual'], output['Prediction'])[0, 1]**2
  print('R2 :', round(R2*100, 2), '%.')

  # Returning the output Dataframe'''
  return output
```

### 1. Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Instantiate model
logreg = LogisticRegression(max_iter=1000)

# Train the model on training data
logreg.fit(training_feature_data, training_Fraudulence)

# Testing the model on the test data using the test target and displaying it
test_output_log = predict(testing_feature_data, testing_Fraudulence, logreg)
test_output_log
```

### 2. Using Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

# Instantiate model with 10 decision trees
rf1 = RandomForestRegressor(n_estimators = 10, random_state = 42)

# Train the model on training data
rf1.fit(training_feature_data, training_Fraudulence)

# Testing the model on the test data using the test target and displaying it
test_output_rf1 = predict(testing_feature_data, testing_Fraudulence, rf1)
test_output_rf1
```

### 3. Using Decision Tree



```python
from sklearn.tree import DecisionTreeRegressor

# Instantiate model
dt = DecisionTreeRegressor(random_state=42)

# Train the model on training data
dt.fit(training_feature_data, training_Fraudulence)

# Testing the model on the test data using the test target and displaying it
test_output_dt = predict(testing_feature_data, testing_Fraudulence, dt)
test_output_dt
```
